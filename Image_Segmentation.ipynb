```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
from skimage import filters
import skimage
```

```python
warnings.filterwarnings("ignore")
```

```python
image = cv2.imread("010obverse.jpg", 0)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
flattened_image = image.reshape((image.shape[0] * image.shape[1],))
```

```python
flattened_image.shape
```

```python
plt.figure()
sns.distplot(flattened_image, kde=True)
plt.show()
```

Threshold based segmentation

```python
ret, thresh1 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(8, 8))
plt.title('Threshold based segmentation')
plt.imshow(thresh1, cmap="binary")
plt.axis("off")
plt.show()
```

**Edge based segmentation**

```python
plt.figure(figsize=(8, 8))
plt.title('Regular Image')
plt.imshow(image)
plt.axis("off")
plt.show()
```

Applying Sobel Edge operator

```python
sobel_image = filters.sobel(image)
```

```python
plt.figure(figsize=(8, 8))
plt.title('Sobel Edge Operator')
plt.imshow(sobel_image)
plt.axis("off")
plt.show()
```

Applying Roberts Edge operator

```python
roberts_image = filters.roberts(image)
```

```python
plt.figure(figsize=(8, 8))
plt.title('Roberts Edge Operator')
plt.imshow(roberts_image)
plt.axis("off")
plt.show()
```

Applying Prewitt edge operator

```python
prewitt_image = filters.prewitt(image)
plt.figure(figsize=(8, 8))
plt.title('Prewitt Edge Operator')
plt.imshow(prewitt_image)
plt.axis("off")
plt.show()
```
